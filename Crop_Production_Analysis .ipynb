**Summary About Dataset**

The dataset contains information about crop production across various states and districts.
Attributes include State_Name, District_Name, Crop_Year, Season, Crop, Area, and Production.

*   Total number of states - `29`
*   Total number of Districts - `512`
*   Duration in year - `1997 to 2015`
*   Total number of entries - `184674`







**Columns Description:**

`1. State_Name` : This Column Contains the name of a states

`2. District_Name` : This contains the name of District

`3. Crop_Year` : This column contain year Ranging from `1997` to `2015`

`4. Season` : This column contain the seasons of agriculture such as `Kharif`, `Whole Year`, `Autumn`, `Rabi`, `Summer`, `Winter`

`5. Crop` : This column contains the name of various crops

`6. Area` : This column indicates the area of land dedicated to cultivating each crop

`7. Production` : This column indicating the amount of each crop harvested

```python
mean = np.mean(Crop['Production'])
std = np.std(Crop['Production'])
```

```python
# Define threshold for outlier detection (e.g., Z > 3 or Z < -3)
threshold = 3
```

```python
# calculate z- score
z_score = (Crop['Production'] - mean)/std
```

```python
# find outlier
outlier = np.abs(z_score) > threshold
```

```python
filtered_data = Crop['Production'][~outlier]
```

```python
filtered_data
```

```python
Crop['Production'] = filtered_data
```

```python
# Assuming 'crop' is your Pandas Series containing production data
production_data = crop['Production']

# Calculate the interquartile range (IQR)
Q1 = production_data.quantile(0.25)
Q3 = production_data.quantile(0.75)
IQR = Q3 - Q1

# Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find potential outliers
outliers = production_data[(production_data < lower_bound) | (production_data > upper_bound)]

print("Potential outliers:")
print(outliers)
outliers.count()
```

```python
# avg production
avg_production = crop.groupby(['State_Name','District_Name'])['Production'].mean().reset_index()
```

```python
min_production =df[df['Production'] == min(df['Production'])]['Production']
max_production =df[df['Production'] == max(df['Production'])]['Production']
max_production.values[0]
min_production.values[0]
```

```python
plt.figure(figsize=(15,15))
sns.barplot(data=df,x='District_Name',y='Production',hue='State_Name')
plt.xlabel('District_Name')
plt.ylabel('Production')
plt.title('Top three district ')
plt.xticks(rotation=90)
plt.ylim(min_production.values[0], max_production.values[0])
plt.savefig('top_three_district.png')
plt.show()
```

```python
df = crop[['Season','Crop','Area','Production']].sort_values(by=['Area','Production'],ascending=[False,False]).reset_index(drop=True)
```

```python
df.head()
```

```python
df1 = pd.DataFrame(df.groupby(['Season','Crop'])['Area'].mean()).reset_index()
df2 = pd.DataFrame(df.groupby(['Season','Crop'])['Production'].mean()).reset_index()
```

```python
heatmap_data1 = df1.pivot(columns='Season', index='Crop', values='Area').fillna(0)
heatmap_data2 = df2.pivot(columns='Season', index='Crop', values='Production').fillna(0)
```

```python
# plotting Heatmap for Average Area
plt.figure(figsize=(11,22))
sns.heatmap(heatmap_data1, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=.5)
plt.title('Heatmap of Crop Cultivation Average Areas by Season')
plt.xlabel('Crop')
plt.ylabel('Season')
plt.savefig('HeatMap_AvgArea.png')
plt.show()
```

```python
# plotting Heatmap for Average Production
plt.figure(figsize=(11,22))
sns.heatmap(heatmap_data2, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=.5)
plt.title('Heatmap of Crop Cultivation Average Productin by Season')
plt.xlabel('Crop')
plt.ylabel('Season')
plt.savefig('HeatMap_AvgProduction.png')
plt.show()
```
